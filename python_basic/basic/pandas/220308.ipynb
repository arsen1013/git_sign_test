2. 함수를 하나 생성한다. 함수는 csv파일을 URL에 있는 csv 파일을 읽어오고 "country" 기준으로 sort, 인덱스 값을 초기화 한다. 
3. 함수 하나 더 생성 -> 인자값을 리스트로 받는다.  이 매개변수를 기준으로 columns 삭제

```python
import pandas as pd 
```

```python
#내가 한 거

class class_1():
    def __init__(self,input_url):
        self.url = input_url
        self.input_a = []
        
    def reading(self):
        self.df = pd.read_csv(self.url)
        self.df.sort_values("Country",inplace=True) #inplace 해주기! 
        self.df.reset_index(drop = True, inplace = True)
        return self.df

    def DelColumns(self,input_ColName):
        self.df2 = class_1(self.url).reading() 
        self.input_a.append(input_ColName)
        return self.df2.drop(self.input_a, axis =1) 
    

class_1('./csv/Sales Records.csv').reading().head(3)

```

```python
class_1('./csv/Sales Records.csv').DelColumns("Region").head(3)
```

```python
#쌤이 한 거 

class class_dt():
    def __init__(self,input_url):
        self.url = input_url
        
    def csv_read(self,inport_sort,input_inplace):
        #value 값은 넣어주되, columns은 값을 넣지 않아도 기본값이 있어서 돌아가는데.  
        self.dt = pd.read_csv(self.url)
        self.dt.sort_values(inport_sort,inplace= input_inplace) #inplace 해주기! 
        self.dt.reset_index(drop = True, inplace = True) #인덱스 초기화 
        return self.dt

    def remove_columns(self,input_list):
        self.dt.drop(input_list, axis =1 ,inplace= True) 
        #input_list는 self를 붙이지 않는 이유는 지역변수가 아니라 매개변수임 
        return self.dt
    

dt = class_dt('./csv/Sales Records.csv')

```

```python
class class_dt():
    def __init__(self,input_url):
        self.url = input_url
        
    def csv_read(self,inport_sort= "Country",input_inplace = True):
        self.dt = pd.read_csv(self.url)
        print(self.dt)
        self.dt.sort_values(inport_sort ,inplace = input_inplace ) 
        self.dt.reset_index(drop = True, inplace = True)  
        return self.dt

    def remove_columns(self,input_list = []):
        #빈 리스트를 기본값으로 들어감 
        self.dt.drop(input_list, axis =1 ,inplace= True) 
        return self.dt

    def remove_2(self, input_s_column, inout_e_column):
        self.dt.drop(self.dt.loc[:, input_s_column : inout_e_column], axis =1, inplace =True )
        return self.dt
    

dt = class_dt('./csv/Sales Records.csv')
```

일일 확진자를 출력하려면 어떻게 계산식을 도출해야할까? shift 함수 사용 


```python
#새로운 컬럼 추가 
corona_df_2["일일 확진자"] = (corona_df_2["확진자"]- corona_df_2["확진자"].shift()).fillna(0)
# fillna() : 값이 NaN인 경우 ()값으로 대체
#shift는 행을 한칸씩 옮겨준다. 
```

```python
corona_df_2
```

```python
corona_df_2["일일 사망자"] = (corona_df_2["사망자"]- corona_df_2["사망자"].shift()).fillna(0)
# fillna() : 값이 NaN인 경우 ()값으로 대체
#diff 와 달리 shift를 쓰는 방법은 계산이 조금 더 복잡해질때 사용한다. 두칸씩 띄어서 계산하거나, 월간을 구하거나 할 때. 

corona_df_2
```

```python
corona_df_2["일일 사망자"] = corona_df_2["사망자"].diff().fillna(0)
#diff 도 바로 전 데이터의 차를 집어넣어줌  
corona_df_2
```

조건을 여러개 출력할때야 말로, 함수를 여러개 쓰는 것이 apply의 꽃 

1부터 ~5 까지 순서대로 넣었는데, 1이랑 2면 0으로 출력하고, 

```python

#원하는 컬럼 하나씩 변경하기

#뭔가의 매개변수를 넣는 함수 만들기 
def change (x): 
    if x < 3: 
        return 0 
    elif x < 5: 
        ##5보다 작다곤 했지만 실질적으로 정수로 따지자면 3,4 만 가능한 것 
        return 1
    elif x < 7: 
        return 2 
    else: 
        return 3 


test.A = test.A.apply(change)
#test["A"]로 적어도 똑같음 
test
```

smith의 연봉을 10% 인상하기 

한사람만 수정하기 
update _ one  

```python
db.emp.update_one(
    {
        "ENAME":"SMITH"
    },{
        "$set":{
            "SAL" : 880
        }
    }
)
```

smith 월급이 수정 됐는지 확인한다. 

```python
result = db.emp.find({
    "ENAME":"SMITH"
})
pd.DataFrame(result)
```

smith 가 3명 있는데 update_one을 하니 한사람만 올려준다. 

여러사람을 수정하려면? 
update_many({},{})

엘런을 수정해본다. 

```python
db.emp.update_many(
    {
        "ENAME":"ALLEN"
    },{
        "$set" : {
            "SAL" : 3200
        }
    }
)
result = db.emp.find({
    "ENAME":"ALLEN"
})
pd.DataFrame(result)
```

```python
db.emp.delete_many(
    {
        "ENAME":"PRESIDENT"
    }
)
result = db.emp.find({
    "JOB":"PRESIDENT"
})
pd.DataFrame(result)
```

파이썬에서 %s , %d, %f 로 리스트에 있는 값을 담아주는 문구다. 
그래서 SQL문을 사용할 수 없다. 
이를 해결하기 위해서는 %%를 사용한다. 

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE 'S%%'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE '%%S'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE '%%S%%'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

____ 는 글자수 만큼 언더바를 써주면 된다. 
글자수가 4개인 스미스와 스캇을 출력해보자면

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE 'S____'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE '%%E_'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

```python
sql_25 = """

            SELECT * FROM emp WHERE ENAME LIKE '%%ES'
        """
result_24 = _dt.executeAll(sql_25)
pd.DataFrame(result_24)
```

이렇게 조건을 하나씩 좁혀가면서 원하는 글자를 찾아낼 수 있다. 

실생활에서 어떨 때 like를 쓸 까? 
1. 고객정보가 있는데 김씨 성만 뽑고싶다. 이씨 성만 뽑고싶다. 
2. 특정 지역 주소만 뽑고 싶을때 서울특별시 강북구만 뽑고싶다면 쓸 수 있다. 
3. 검색엔진에서 단어 하나 적으면 연관된 검색어가 다 나온다.

gorup by 를 쓰면  뭔가를 선택을 해서 묶어주기는 하는데, 나머지 컬럼의 데이터 값들을 어떻게 해줄 수가 없다. 

만개의 데이터가 있는데 이것을 국가별로 묶었다고 생각을 해보자 
국가별로 묶으면, 나머지 값들은 int만 있으면 합쳐지겠지만 문자열은 합쳐져도 되는게 아니다. 그래서 그룹바이만 쓰면 홀로 작용하지 않는다. 

그래서 그룹바이 앞에 어떤 컬럼을 기준으로 합칠 것인지 설정해 줘야 한다. 

```python
sql_1 ="""
        SELECT * FROM sales GROUP BY Country
        """
result_1 = _db.executeAll(sql_1) #select는 all로 불러온다! 
df_2 = pd.DataFrame(result_1)
df_2
```

나오긴 하는데,, 뭐가 합쳐진건지 도통 알 수 없다. 

```python
sql_3 = """
        SELECT country, SUM('Unit Sold') FROM sales GROUP BY country
        """
result_3 = _db.executeAll(sql_3)
df_3 = pd.DataFrame(result_3)
df_3
```

보기 나쁜 SUM unit sold라는 출력 컬럼을 unit sold로 바꿀 수 있다. 

```python
sql_3 = """
        SELECT country, SUM('Units Sold') AS 'Unit Sold'
         FROM sales GROUP BY country
        """
result_3 = _db.executeAll(sql_3)
df_3 = pd.DataFrame(result_3)
df_3
```

 ' ' 가 아니라  `` 를 써야 한다. 
 따옴표가 아니라 백쿼트를 써야 한다. 
 컬럼 값에 공백이 있는데, 공백같은 경우 백쿼트를 넣으면 공백도 문자열로 취급해서 같이 들어간다. 

 백쿼트로 스트링을 처리할 때가 있다. 

 보통 대부분의 언어에서 """ """ 대신 ` ` 를 쓴다. 

```python
sql_4 = """
        SELECT country, SUM(`Units Sold`) AS 'Unit Sold' 
        FROM sales GROUP BY country
        """
result_4 = _db.executeAll(sql_4)
df_4 = pd.DataFrame(result_4)
df_4
```

```python
sql_3 = """
        SELECT country, 
        SUM(`Units Sold`) AS 'Unit Sold'
        FROM sales 
        GROUP BY country
        ORDER BY `Unit Sold`
        """
result_4 = _db.executeAll(sql_3)
df_4 = pd.DataFrame(result_4)
df_4
```

판다스에서는 순서 정렬하고 reset index 해주는 방법과 
이렇게 SQL로 한번에 인덱스 초기화 까지 하는 방법이 있다. 

sql이 조금더 속도가 빠르지 않을까 싶다. 


```python
sql_5 = """
        SELECT country, 
        SUM(`Units Sold`) AS 'Unit Sold'
        FROM sales 
        GROUP BY country
        ORDER BY `Country` DESC
        """
result_5 = _db.executeAll(sql_5)
df_5 = pd.DataFrame(result_5)
df_5
```

```python
#AVG 평균 합산 

sql_6 = """
        SELECT country, 
        AVG(`Units Sold`) AS 'Unit Sold'
        FROM sales 
        GROUP BY country
        ORDER BY `Country` DESC
        """
result_6 = _db.executeAll(sql_6)
df_6 = pd.DataFrame(result_6)
df_6
```
